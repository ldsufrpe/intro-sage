## Coordenadas Parametricas

```python
t = var('t')
parametric_plot((e^t*cos(2*pi*t), e^t*sin(2*pi*t)), (t,0,4)) ###Espiral Logaritmica
```

## Coordenadas Polares

```python
polar_plot(e^(t/(2*pi)), (t,0,8*pi)) ##Mesma Curva Anterior em coordenadas polares.
```

```python
polar_plot(sin(15*x)^2, (x, 0, 2*pi)) #rosacea de 30 petalas
```

## Curvas Definidas Implicitamente

**Elipse.**

```python
x, y = var('x, y')
```

```python
implicit_plot(x^2/4+y^2/9==1, (x,0,1), (y,-3,3)) ## Atencao aos valores de x
```

```python
implicit_plot(x^2/4+y^2/9==1, (x,-2,2), (y,-3,3)) ## Atençao aos valores de x
```

Elipse em coordenadas polares

```python
b = 3
a = 2
# e=c/a=sqrt(b^2-a^2)/b
e = sqrt(b^2-a^2)/b
```

```python
polar_plot(b*(1-e^2)/(1+e*cos(x)), (x, 0, 2*pi)) ###Falta Transladar e Girar.(o Foco esta na origem)
```

**Customizando as plotagens.**

```python
G1 = plot(2*exp((-x^4)), (-1.5,1.5), linestyle='solid')
G2 = plot(3*exp((-x^4)), (-1.5,1.5), linestyle='dashed')
G3 = plot(4*exp((-x^4)), (-1.5,1.5), linestyle='dotted')
(G1 + G2 + G3).show()
```

```python
G1 = plot(2*exp((-x^4)), (-1.5,1.5), thickness=1)
G2 = plot(3*exp((-x^4)), (-1.5,1.5), thickness=3)
G3 = plot(4*exp((-x^4)), (-1.5,1.5), thickness=6)
(G1 + G2 + G3).show()
```

```python
plot(sin(x), (0,4*pi), legend_label='y=sin(x)',title='Gráfico de f(x)=sin(x)')
```

Preenchimentos

```python
p1 = plot(sin(x), (x, 0, 4*pi), fill='axis')
p2 = plot(sin(x), (x, 0, 4*pi), fill='min',fillalpha=1.0)
p3 = plot(sin(x), (x, 0, 4*pi), fill='max', fillalpha=0.25)
p4 = plot(sin(x), (x, 0, 4*pi), fill=x,fillcolor='red', fillalpha=0.3)
G = graphics_array([[p1, p2], [p3, p4]])
G.show(figsize=[7,5], aspect_ratio=1)
```
